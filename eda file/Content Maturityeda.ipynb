```python
import pandas as pd
import matplotlib.pyplot as plt

```

# Load dataset


```python
df = pd.read_csv(r"C:\Users\admin\Downloads\Atlantic_Spain.csv")
```

# Convert date


```python
df['date'] = pd.to_datetime(df['date'])
```

# Basic Info


```python
print("Dataset Shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nMissing Values:\n", df.isnull().sum())
```

# Songs per Day (Data Check)


```python
songs_per_day = df.groupby('date').size()

plt.figure()
songs_per_day.plot()
plt.title("Songs per Day (Should be 50)")
plt.xlabel("Date")
plt.ylabel("Count")
plt.show()
```

# Top 10 Artists (Most Frequent)


```python
top_artists = df['artist'].value_counts().head(10)

plt.figure()
top_artists.plot(kind='bar')
plt.title("Top 10 Artists")
plt.xlabel("Artist")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# Position Distribution

```python

plt.figure()
df['position'].plot(kind='hist', bins=20)
plt.title("Position Distribution")
plt.xlabel("Position")
plt.ylabel("Frequency")
plt.show()
```

# Explicit vs Non-Explicit

```python
explicit_counts = df['is_explicit'].value_counts()

plt.figure()
explicit_counts.plot(kind='bar')
plt.title("Explicit vs Clean Songs")
plt.xlabel("Explicit (True/False)")
plt.ylabel("Count")
plt.show()
```

# Album Type Distribution

```python
album_counts = df['album_type'].value_counts()

plt.figure()
album_counts.plot(kind='bar')
plt.title("Album Type Distribution")
plt.xlabel("Album Type")
plt.ylabel("Count")
plt.show()
```

# Popularity vs Position

```python
plt.figure()
plt.scatter(df['position'], df['popularity'])
plt.title("Position vs Popularity")
plt.xlabel("Position")
plt.ylabel("Popularity")
plt.show()
```

# Song Duration Distribution


```python
df['duration_min'] = df['duration_ms'] / 60000

plt.figure()
df['duration_min'].plot(kind='hist', bins=20)
plt.title("Song Duration (Minutes)")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()
```

# Daily Average Popularity


```python
daily_pop = df.groupby('date')['popularity'].mean()

plt.figure()
daily_pop.plot()
plt.title("Daily Average Popularity")
plt.xlabel("Date")
plt.ylabel("Popularity")
plt.show()
```
